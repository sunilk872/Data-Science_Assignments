```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

# Split the raw data into correct form.

Part 1: Load the Data and Display Initial Information
Python


```python
# Load the Online Retail dataset
online_retail = pd.read_csv('/content/Online retail.csv', header=None)

# Display the first few rows and the shape of the dataset
display(online_retail.head())
print(online_retail.shape)
```

Part 2: Split the First Row

```python
# Split the first row into separate columns
first_row_split = online_retail.iloc[0, 0].split(',')

# Create a DataFrame for the first row
first_row_df = pd.DataFrame([first_row_split])
first_row_df
```

Part 3: Split the Remaining Rows into Separate Columns
Python

```python
# Split the remaining rows into separate columns
remaining_rows_split = online_retail.iloc[:, 0].str.split(',', expand=True)
remaining_rows_split.head(3)
```

Part 4: Ensure Consistent Number of Columns
Python

```python
# Determine the maximum number of columns needed
max_columns = max(first_row_df.shape[1], remaining_rows_split.shape[1])

# Ensure both DataFrames have the same number of columns
first_row_df = first_row_df.reindex(columns=range(max_columns))
remaining_rows_split = remaining_rows_split.reindex(columns=range(max_columns))
```

Part 5: Combine the first row and Remaining Rows
Python

```python
# Combine the first row DataFrame with the remaining rows DataFrame
df = pd.concat([first_row_df, remaining_rows_split], ignore_index=True)

# Fill NaN values with empty strings or any other placeholder if needed
df =df.fillna('')
```

 Remove Null and Duplicate Values

```python
# Remove rows with any null values
df = df.dropna()

# Remove duplicate rows
df = df.drop_duplicates()
```

 Reset the Index

```python
df= df.reset_index(drop=True)
df
```

```python
#Set the first row as the column headers
df.columns = df.iloc[0]

#Drop the first row
df = df.drop(df.index[0])

#Reset the index, starting from 0
df = df.reset_index(drop=True)

# Display the transformed DataFrame
df
```

#Apriori Algorithm

**Convert Categorical Data to Numeric Data**

```python
# Perform get_dummies to convert categorical data to numeric form
dummies_df = pd.get_dummies(df, prefix='', prefix_sep='', dummy_na=True, dtype='int')

# Display the DataFrame with dummy variables
display(dummies_df)
```

**Perform Apriori Algorithm**

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Perform the Apriori algorithm
frequent_itemsets = apriori(dummies_df, min_support=0.3, use_colnames=True)

# Display the frequent itemsets
display(frequent_itemsets)
```

**Generate Association Rules**

```python
# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)

# Display the association rules
rules
```

**Sort Rules by Lift**

```python
# Sort the rules by lift and display the top 20
top_rules = rules.sort_values('lift', ascending=False).head(20)

# Display the top 20 rules
top_rules
```

**Sort Rules by Lift**

```python
rules[rules.lift>1]
```

**Interview Questions:**
1.	What is lift and why is it important in Association rules?
2.	What is support and Confidence. How do you calculate them?
3.	What are some limitations or challenges of Association rules mining?


1. **Lift**: Measures the strength of a rule. Lift > 1 indicates a strong association. It's important because it shows how much more often the antecedent and consequent occur together than expected by chance.

2. **Support and Confidence**:
   - **Support**: Proportion of transactions containing both items.
     $$
     Confidence(X => Y) = (Number of transactions containing X and Y) / (Total number of transactions)
     $$
   - **Confidence**: Proportion of transactions with the antecedent that also have the consequent.
     Confidence(X => Y) = (Number of transactions containing X and Y) / (Number of transactions containing X)

3. **Limitations**:
   - **Scalability**: Large datasets are hard to process.
   - **Redundancy**: Many irrelevant rules.
   - **Interpretability**: Difficult to understand many rules.
   - **Sparsity**: Hard to find meaningful associations in large itemsets.
