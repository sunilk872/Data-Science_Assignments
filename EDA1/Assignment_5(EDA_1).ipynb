```python
import pandas as pd

# Function to detect and remove outliers
def remove_outliers(df):
    for column in df.columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Remove outliers
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    return df

# Remove outliers from the dataset
data_cleaned = remove_outliers(df)

# Display the shape of the data before and after removing outliers
data_shape_before = df.shape
data_shape_after = data_cleaned.shape

data_shape_before , data_shape_after
```

**4.	Pattern Recognition and Insights**

```python
import pandas as pd

# Assuming df is the DataFrame containing the loaded data
correlations = df.corr()

# Identify significant correlations (absolute value > 0.3)
significant_correlations = correlations[(correlations.abs() > 0.3) & (correlations != 1.0)]

# Remove rows and columns with all NaN values (if any)
significant_correlations = significant_correlations.dropna(how='all', axis=0).dropna(how='all', axis=1)

print("Significant Correlations (|correlation| > 0.3):\n")

# Loop through the correlation matrix to print significant correlations
for i in range(len(significant_correlations)):
    for j in range(i + 1, len(significant_correlations)):
        corr_value = significant_correlations.iloc[i, j]

        if abs(corr_value) > 0.3:  # Threshold for significance
            var1 = significant_correlations.index[i]
            var2 = significant_correlations.columns[j]
            print(f"Correlation between {var1} and {var2}: {corr_value:.2f}")

            # Provide simple implications
            if 'NSP' in (var1, var2):
                print(f"  Implication: The relationship between {var1} and {var2} may affect fetal state categories (NSP).\n")
            else:
                print(f"  Implication: {var1} and {var2} might be physiologically connected.\n")
```
